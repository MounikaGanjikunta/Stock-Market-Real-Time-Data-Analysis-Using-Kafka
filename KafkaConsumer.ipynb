```python
from s3fs import S3FileSystem
```

```python
consumer = KafkaConsumer(
    'demo_test',
     bootstrap_servers=['18.117.238.62:9092'], 
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
for c in consumer:
    print(c.value)
```

```python
s3 = S3FileSystem()
```

```python
for count, i in enumerate(consumer):
    with s3.open("s3://kafka-stock-market-mg/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value, file)
```
